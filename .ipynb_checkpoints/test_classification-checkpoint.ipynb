```python
import pandas as pd
import numpy as np
from numpy import zeros
```

```python
#sklearn imports
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
from sklearn.feature_extraction.text import CountVectorizer
from sklearn import preprocessing
from sklearn import cross_validation
from sklearn.metrics import classification_report

from sklearn.model_selection import cross_val_predict
from sklearn.model_selection import cross_val_score
```

```python
#load data
data = pd.read_csv('train_set.csv', sep="\t")
```

```python
#create set from categories
le = preprocessing.LabelEncoder()
le.fit(data["Category"])
y = le.transform(data["Category"])
set(y)
```

```python
y
```

```python
#import pipeline
from sklearn.pipeline import make_pipeline
#import model
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
```

```python
model=make_pipeline(preprocessing.StandardScaler(), KNeighborsClassifier(n_neighbors=3,weights='distance', algorithm='ball_tree',n_jobs=-1))
```

```python
#import vectorizer and lsi
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
from sklearn.decomposition import NMF
```

```python
#do vectorization
vectorizer = TfidfVectorizer(max_df=0.5,stop_words='english',use_idf=True)
X = vectorizer.fit_transform(data['Content'])
lsa=TruncatedSVD(n_components=100)
X=lsa.fit_transform(X)
X=preprocessing.Normalizer(copy=False).fit_transform(X)
```

```python
#model.fit(X,y)
```

```python
#run model with 10_fold
predicted= cross_val_predict(model, X, y, cv=10)
```
